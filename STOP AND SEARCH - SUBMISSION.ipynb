```python
data_november.drop(['operation_name', 'self_defined_ethnicity', 'removal_of_more_than_outer_clothing', 'outcome_linked_to_object_of_search',
              'location.longitude', 'location.latitude', 'location', 'operation', 'location.street.id', 'location.street.name'], 
                   axis=1, inplace=True)
```

```python
data_november.head()
```

```python
#changing datetime to pd.datetime
data_november['datetime'] = pd.to_datetime(data_november['datetime'])
data_november['datetime'] = data_november['datetime'].dt.date
```

```python
data_november.info()
```

```python
#renaming columns
data_november.rename({'officer_defined_ethnicity':'ethnicity', 'datetime':'date'}, axis=1, inplace=True)
```

```python
data_november.head()
```

```python
#checking for the null/missing values
data_november.isnull().sum()
```

```python
#removing the NAs in age_range and ethnicity

data_november_new = data_november.dropna()
```

```python
data_november_new.info()
```

```python
#visualizing data by the month considered using date

plot_month_november = data_november_new.groupby('date').size().reset_index(name='number of outcomes').set_index('date')
plot_month_november.plot(kind='line', figsize=(20,10))
```

```python
#visualizing date by the object of search using groupby of date and obejct_of_search

count_by_date = data_november_new.groupby('date').size()

plt.figure(figsize=(20,10))
plt.xlabel('date')
plt.ylabel('object_of_search')
plt.title('date and search ventures')
plt.plot(count_by_date)
```

```python
#visualizing the trend of stop and search for the month for gender

gender_count = data_november_new.groupby('date')
gender_count = gender_count.gender.apply(pd.value_counts).unstack(-1).fillna(0)

gender_count.plot(kind='line', figsize=(15,10), title='trend of stop and search count of gender')
```

```python
#visualizing the trend of stop and search for the month by ethnicity

ethnicity_count = data_november_new.groupby('date')
ethnicity_count = ethnicity_count.ethnicity.apply(pd.value_counts).unstack(-1).fillna(0)

ethnicity_count.plot(kind='line', figsize=(15,10), title='trend of stop and search count by ethnicity')
```

```python
#visualizing the trend of stop and search for the month by age range

age_count = data_november_new.groupby('date')
age_count = age_count.age_range.apply(pd.value_counts).unstack(-1).fillna(0)

age_count.plot(kind='line', figsize=(15,10), title='trend of stop and search count by age range')
```

```python
#visualizing the trend of stop and search for the month by object of search

object_count = data_november_new.groupby('date')
object_count = object_count.object_of_search.apply(pd.value_counts).unstack(-1).fillna(0)

object_count.plot(kind='line', figsize=(15,10), title='trend of stop and search count by object of search')
```

```python
#visualizing the trend of stop and search by type of stop and search

type_count = data_november_new.groupby('date')
type_count = type_count.type.apply(pd.value_counts).unstack(-1).fillna(0)

type_count.plot(kind='line', figsize=(15,10), title='trend of stop and search type')
```
